These are the tables used in the curated layer of our pipeline ("silver layer", in medallion terms).

The merge tables are overwritten in each import batch and then upserted into the 'real' tables through the merge Python function.

```python
import duckdb
```

```python
con = duckdb.connect('/workspace/data/pipeline.duckdb')
```

```python
con.execute("CREATE SCHEMA IF NOT EXISTS curated")
```

```python
con.execute(""" 
            CREATE OR REPLACE TABLE curated.anime(
            raw_id INTEGER NOT NULL,
            batch_id INTEGER,
            page INTEGER,
            anime_id INTEGER PRIMARY KEY,
            title STRING,
            title_english STRING,
            type STRING,
            source STRING,
            no_of_episodes INTEGER,
            status STRING,
            duration STRING,
            score FLOAT,
            scored_by INTEGER,
            rank INTEGER,
            popularity INTEGER,
            members INTEGER,
            favorites INTEGER,
            year INTEGER
            )
            """)

con.execute(""" 
            CREATE OR REPLACE TABLE curated.anime_merge(
            raw_id INTEGER NOT NULL,
            batch_id INTEGER,
            page INTEGER,
            anime_id INTEGER PRIMARY KEY,
            title STRING,
            title_english STRING,
            type STRING,
            source STRING,
            no_of_episodes INTEGER,
            status STRING,
            duration STRING,
            score FLOAT,
            scored_by INTEGER,
            rank INTEGER,
            popularity INTEGER,
            members INTEGER,
            favorites INTEGER,
            year INTEGER
            )
            """)
```

```python
con.execute(""" 
            CREATE OR REPLACE TABLE curated.links_anime_genres(
            anime_id INTEGER,
            genre_id INTEGER,
            genre_name STRING,
            PRIMARY KEY (anime_id, genre_id)
            )
            """)

con.execute(""" 
            CREATE OR REPLACE TABLE curated.links_anime_genres_merge(
            anime_id INTEGER,
            genre_id INTEGER,
            genre_name STRING,
            PRIMARY KEY (anime_id, genre_id)
            )
            """)
```
