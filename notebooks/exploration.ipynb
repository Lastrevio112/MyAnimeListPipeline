This code is for trying to figure out how many animes are above a certain number of members. This is for the purposes of designing the data architecture - figuring out how many animes have over 10 000 members so that I don't bother extracting obscure anime.

```python
for page in [1, 10, 50, 100, 200, 300]:
    response = httpx.get(f"https://api.jikan.moe/v4/anime?order_by=members&sort=desc&limit=25&page={page}")
    data = response.json()
    last_anime = data['data'][-1]
    print(f"Page {page} | Last anime: {last_anime['title']} | Members: {last_anime['members']:,}")
    time.sleep(1)
```
